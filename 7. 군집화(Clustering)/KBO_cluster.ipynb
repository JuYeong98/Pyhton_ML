```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
KBO_df = pd.read_csv('Pre_Season_Batter.csv')
KBO_df
```

```python
KBO_df['height/weight']
weight=[]
for i in KBO_df['height/weight']:
    weight2 = i[6:9]
    if 'k' in weight2:
        weight2 = weight2[0:2]
    weight.append(weight2)        
```

```python
KBO_df['weight'] = weight
#KBO_df.head(3)
Batter_hit_df  = pd.DataFrame(KBO_df['HR'])
Batter_hit_df['weight'] = KBO_df['weight']
Batter_hit_df.head

```

```python
Batter_hit_df.groupby('weight')['HR'].agg([max,min])
```

```python
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
%matplotlib inline
#kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300,random_state=0).fit(Batter_hit_df)

inertia_arr =[]
k_range = range(2,15)

for k in k_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300,random_state=0).fit(Batter_hit_df)
    inertia  = kmeans.inertia_
    print('k: ', k , 'inertia : ', inertia)
    
    inertia_arr.append(inertia)
print(inertia_arr)

plt.plot(k_range , inertia_arr)
plt.vlines(3, ymin = min(inertia_arr)*0.9999, ymax = max(inertia_arr)*1.003, linestyles = '--', colors='g')
plt.vlines(4, ymin = min(inertia_arr)*0.9999, ymax = max(inertia_arr)*1.003, linestyles = '--', colors='r')

plt.title('KBO_Kmeans')
plt.xlabel('num of clusters')
plt.ylabel('value of inertia')
plt.show()
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300,random_state=0).fit(Batter_hit_df)
```

```python
# KMeans 객체를 이용하여 X 데이터를 K-Means 클러스터링 수행 
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=200, random_state=0)
cluster_labels = kmeans.fit_predict(Batter_hit_df)
Batter_hit_df['kmeans_label']  = cluster_labels

#cluster_centers_ 는 개별 클러스터의 중심 위치 좌표 시각화를 위해 추출
centers = kmeans.cluster_centers_
unique_labels = np.unique(cluster_labels)
markers=['o', 's', '^', 'P','D','H','x']

# 군집된 label 유형별로 iteration 하면서 marker 별로 scatter plot 수행. 
for label in unique_labels:
    label_cluster = Batter_hit_df[Batter_hit_df['kmeans_label']==label]
    center_x_y = centers[label]
    plt.scatter(x=label_cluster['HR'], y=label_cluster['weight'], edgecolor='k', 
                marker=markers[label] )
    
    # 군집별 중심 위치 좌표 시각화 
    plt.scatter(x=center_x_y[0], y=center_x_y[1], s=200, color='white',
                alpha=0.9, edgecolor='k', marker=markers[label])
    plt.scatter(x=center_x_y[0], y=center_x_y[1], s=70, color='k', edgecolor='k', 
                marker='$%d$' % label)

plt.show()
```

```python
Batter_hit_df['cluster']=kmeans.labels_
KBO_result = Batter_hit_df.groupby(['cluster'])['weight'].count()
print(KBO_result)
Batter_hit_df.head(100)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(Batter_hit_df)

Batter_hit_df['pca_x'] = pca_transformed[:,0]
Batter_hit_df['pca_y'] = pca_transformed[:,1]
Batter_hit_df.head(100)
```

```python
# cluster 값이 0, 1, 2 인 경우마다 별도의 Index로 추출
marker0_ind = Batter_hit_df[Batter_hit_df['cluster']==0].index
marker1_ind = Batter_hit_df[Batter_hit_df['cluster']==1].index
marker2_ind = Batter_hit_df[Batter_hit_df['cluster']==2].index

# cluster값 0, 1, 2에 해당하는 Index로 각 cluster 레벨의 pca_x, pca_y 값 추출. o, s, ^ 로 marker 표시
plt.scatter(x=Batter_hit_df.loc[marker0_ind,'pca_x'], y=Batter_hit_df.loc[marker0_ind,'pca_y'], marker='o') 
plt.scatter(x=Batter_hit_df.loc[marker1_ind,'pca_x'], y=Batter_hit_df.loc[marker1_ind,'pca_y'], marker='s')
plt.scatter(x=Batter_hit_df.loc[marker2_ind,'pca_x'], y=Batter_hit_df.loc[marker2_ind,'pca_y'], marker='^')

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('3 Clusters Visualization by 2 PCA Components')
plt.show()
```
